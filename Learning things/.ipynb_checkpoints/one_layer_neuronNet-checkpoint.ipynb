```python
import numpy as np
```

```python
#handles large sigmoid values so it doesnt overflow
from scipy.special import expit


class one_layer_NeuronNetwork:

    def __init__(self, feature_size, hidden_size, learning_rate=0.01):

        self.learning_rate = learning_rate
        
        self.weights = {
           "W1": np.random.randn(hidden_size, feature_size) * 0.01,
           "b1": np.zeros((hidden_size, 1)),
           "W2": np.random.randn(1, hidden_size) * 0.01,
           "b2": np.zeros((1, 1)),
       }


    def forward_prop(self, X):
        
        W1 = self.weights["W1"]
        W2 = self.weights["W2"]
        b1 = self.weights["b1"]
        b2 = self.weights["b2"]

        Z1 = W1 @ X + b1 #(hidden_size, m)
        A1 = np.maximum(0, Z1) #(hidden_size,m)
 
        Z2 = W2 @ A1 + b2 #(1, m)
        A2 = expit(Z2)

        return (Z1, A1, Z2, A2), A2

    def predict(self, X):

        _, A2 = self.forward_prop(X)

        predictions = (A2 > 0.5).astype(int) 

        return predictions

    def compute_derivatives_for_batch(self, X, Y):

        m = X.shape[1] #num training examples

        W1 = self.weights["W1"]
        W2 = self.weights["W2"]
        b1 = self.weights["b1"]
        b2 = self.weights["b2"]

        (Z1, A1, Z2, A2), _ = self.forward_prop(X)
        
        delta = A2 - Y # (1, m)
        
        db2 = (1 / m) * np.sum(delta, axis = 1, keepdims=True) #(1, 1)
        dW2 = (1 / m) * delta @ A1.T # (1, hidden_size)

        dA1 = W2.T @ delta # (hidden_size, m)
        dZ1 = dA1 * (Z1 > 0) # (hidden_size, m)

        db1 = (1 / m) * np.sum(dZ1, axis = 1, keepdims=True) # (hidden_size, 1)
        dW1 = (1 / m) * dZ1 @ X.T # (hidden_size, feature_size)

        grads = {

            "dW1": dW1, # (hidden_size, feature_size)
            "db1": db1, # (hidden_size, 1)
            "dW2": dW2, # (1, hidden_size)
            "db2": db2  # (1, 1)
             
        }

        return grads

    def train_on_batch(self, X, Y):

        grads = self.compute_derivatives_for_batch(X, Y)

        for key in grads:
            #they have to have the same name! just letter d difference
            self.weights[key[1:]] -= self.learning_rate * grads[key]
```

```python
import h5py

def load_data():
    
    train_dataset = h5py.File('train_catvsnoncat.h5', "r")
    test_dataset = h5py.File('test_catvsnoncat.h5', "r")

    X_train = np.array(train_dataset["train_set_x"])  # (209, 64, 64, 3)
    y_train = np.array(train_dataset["train_set_y"])  # (209,)

    X_test = np.array(test_dataset["test_set_x"])     # (50, 64, 64, 3)
    y_test = np.array(test_dataset["test_set_y"])     # (50,)

    num_examples = X_train.shape[0]
    feature_size = X_train.shape[1] * X_train.shape[2] * X_train.shape[3] 

    # Flatten and normalize
    X_train = X_train.reshape(feature_size, num_examples) / 255.0 # (12288, 209)

    #shortcut:
    X_test = X_test.reshape(X_test.shape[0], -1).T / 255.0     # (12288, 50)
    y_train = y_train.reshape(1, -1)  # (1, 209)
    y_test = y_test.reshape(1, -1)    # (1, 50)

    return X_train, y_train, X_test, y_test
```

```python
X_train, y_train, X_test, y_test = load_data()

NN = one_layer_NeuronNetwork(X_train.shape[0], 10, 0.001)
```

```python
def compute_cost(yhat, y_train):

    m = y_train.shape[1]

    losses = np.multiply(y_train, np.log(yhat)) + np.multiply((1 - y_train), np.log(1 - yhat)) # (1, 209)
    cost = - (1 / m) * np.sum(losses, axis = 1, keepdims=True) # (1, 1)
    
    cost = float(np.squeeze(cost))  # convert [[value]] → value

    return cost
```

```python
accuracies = []
costs = []
```

```python
from tqdm import tqdm

epochs = 10000

progress = tqdm(range(epochs), desc="Training", leave=True)

for epoch in progress:

    NN.train_on_batch(X_train, y_train)

    if epoch % 100 == 0:

        predictions = NN.predict(X_train) # (1, 209)
        accuracy = np.mean(predictions == y_train) # (1, 209) == (1, 209)

        _, yhat = NN.forward_prop(X_train)
        
        cost = compute_cost(yhat, y_train)

        accuracies.append((epoch, accuracy))  # Store for plotting
        costs.append((epoch, cost)) # Store for plotting

        progress.set_postfix({
            
            "Accuracy": f"{accuracy * 100:.2f}%",
            "Cost": f"{cost}"
        
            })
```

```python
import matplotlib.pyplot as plt

# Split stored values
epochs_list, acc_values = zip(*accuracies)

plt.figure(figsize=(8, 5))
plt.plot(epochs_list, [a * 100 for a in acc_values])
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Training Accuracy Over Time")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Split stored values
epochs_list, cost_values = zip(*costs)

plt.figure(figsize=(8, 5))
plt.plot(epochs_list, cost_values)
plt.xlabel("Epoch")
plt.ylabel("Cost")
plt.title("Cost Over Time")
plt.grid(True)
plt.show()
```

```python
predictions = NN.predict(X_test)
accuracy = np.mean(predictions == y_test)

print(f"Accuracy of test set: {accuracy * 100:.2f}%")
```
